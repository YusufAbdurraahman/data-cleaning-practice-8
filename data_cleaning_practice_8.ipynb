# **Data Cleaning**

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
threshold = 0.75 * len(df)

less = missing_values[missing_values < threshold].index
over = missing_values[missing_values >= threshold].index

less, over, threshold
```

```python
numeric_features = df[less].select_dtypes(include = ["number"]).columns
df[numeric_features] = df[numeric_features].fillna(df[numeric_features].median())
```

```python
kategorikal_feature = df.select_dtypes(include = ["object"]).columns

for columns in kategorikal_feature:
  df[columns] = df[columns].fillna(df[columns].mode()[0])
```

```python
df = df.drop(columns = over)
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
for feature in numeric_features:
  plt.figure(figsize = (12,5))
  sns.boxplot(y = df[feature])
```
